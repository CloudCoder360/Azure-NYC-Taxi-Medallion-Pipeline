# Data Reading and Writing and Creating Delta Tables(External)

**Variable Assingning**

```python
silver = 'abfss://silver@nyctaxistorageshazal.dfs.core.windows.net'
gold = 'abfss://gold@nyctaxistorageshazal.dfs.core.windows.net'

```

### Data Zone

```python
df_trip_zone = spark.read.format('parquet')\
                     .option("inferschema", "true").option('header' , True)\
                     .load(f'{silver}/trip_zone')
```

```python
df_trip_zone.display()
```

# DataBase Creation

```python
%sql
CREATE DATABASE gold
```

### Creating External Delta Tables of parquet files

**Trip Zone**

```python
df_trip_zone.write.format('delta').mode('append').option(
    'path', f'{gold}/trip_zone'
).saveAsTable('gold.trip_zone')

```

```python
%sql
select * from gold.trip_zone
where Borough = 'EWR'
```

**Trip Type**

```python
df_trip_type = spark.read.format('parquet')\
                     .option("inferschema", "true").option('header' , True)\
                     .load(f'{silver}/trip_type')
```

```python
df_trip_type.display()
```

```python
df_trip_type.write.format('delta').mode('append').option(
    'path', f'{gold}/trip_type'
).saveAsTable('gold.trip_type')
```

**Trip Data**

```python
df_trip_data = spark.read.format('parquet')\
                     .option("inferschema", "true").option('header' , True)\
                     .load(f'{silver}/trip_data')
```

```python
df_trip_data.display()
```

```python
df_trip_data.write.format('delta').mode('append').option(
    'path', f'{gold}/trip_data'
).saveAsTable('gold.trip_data')
```

# Learning Delta Lake

###  Versioning

```python
%sql
select * from gold.trip_zone
where LocationID = 1
```

```python
%sql

UPDATE gold.trip_zone
SET Borough = 'EMR'
WHERE LocationID = 1
```

```python
%sql
delete from gold.trip_zone
where LocationID = 1
```

**Track Versioning**

```python
%sql
describe history gold.trip_zone
```

### Time Travel

```python
%sql
restore gold.trip_zone to version as of 0;
    
```
